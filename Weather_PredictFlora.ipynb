***Apply an AI model called Ridge regression model***

```python
from sklearn.linear_model import Ridge

rr = Ridge(alpha=.1)
```

***We will decide the columns that will be used for our predictions***

```python
predictors = weather.columns[~weather.columns.isin(["PREDICTTMAX", "NAME", "STATION"])]
```

```python
predictors
```

In the function we are asking our model to take past 10yrs of data and predict it for next 90 days with a rolling 90 days pattern. We will give our actual data **"PREDICTTMAX"** to the model to predict a value that will  ***fit into that value***image.png

```python
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        model.fit(train[predictors], train["PREDICTTMAX"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["PREDICTTMAX"], preds], axis=1)
        combined.columns = ["ACTUAL", "PREDICTION"]
        combined["DIFF"] = (combined["PREDICTION"] - combined["ACTUAL"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors)
```

***Check the predictions***

```python
predictions
```

***Find the mean error from the difference.*** This is an ***accuracy metirc*** to evaluate the ***model effectiveness***. Mean error is the average of all the diffs in the predictions above

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mean_absolute_error(predictions["ACTUAL"], predictions["PREDICTION"])
```

```python
def pct_diff(old, new):
    return (new - old) / old

def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    weather[label] = weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather

rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["TMAX", "TMIN", "PRCP"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["TMAX", "TMIN", "PRCP"]:
    weather[f"month_avg_{col}"] = weather[col].groupby(weather.index.month, group_keys=False).apply(expand_mean)
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys=False).apply(expand_mean)
```

```python
weather = weather.iloc[14:,:]
weather = weather.fillna(0)
```

```python
predictors = weather.columns[~weather.columns.isin(["PREDICTTMAX", "NAME", "STATION"])]
```
